```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime
```

```python
home_data = pd.read_excel('HomePricebyState.xlsx', index_col=0, header=0,sheet_name='Sheet3')
```

```python
home_data = home_data.sort_values(by=['2017-03-01 00:00:00'],ascending=False)
home_data.head()
```

```python
index = home_data.columns
index
```

```python
i=1
plt.figure(figsize=(16,10))
plt.title('Home Price Increase since 1996 by State')
plt.xlabel("Year")
plt.ylabel("Growth (indexed to 1996 = 1)")
plt.figtext(.6,.82,"*Note:\n Legend sorted by growth", wrap=True,
            horizontalalignment='left', fontsize=12)

for i in range(51):
    plt.plot(index,home_data.iloc[i],label=home_data.index[i])

plt.plot(index, home_data.iloc[19], marker='', color='black', linewidth=4, alpha=0.7,linestyle='--')
plt.legend(loc='upper left',ncol=3)


```
